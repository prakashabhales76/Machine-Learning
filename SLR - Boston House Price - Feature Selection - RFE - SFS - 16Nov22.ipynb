```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
y = boston.target
bos = pd.DataFrame(boston.data, columns=boston.feature_names)
bos['price'] = y
```

```python
from sklearn.feature_selection import RFE
```

```python
y = bos['price']
X = bos.drop('price', axis=1)
```

```python
cols=list(X.columns)
cols.remove('CHAS')
for col in cols:
    X[col+'_2'] = X[col]**2
```

```python
X.columns
```

```python
from sklearn.linear_model import LinearRegression
lir = LinearRegression()
rfe = RFE(lir, n_features_to_select=10, verbose=2)
rfe.fit(X, y)
```

```python
rfe.ranking_
```

```python
pd.DataFrame(rfe.ranking_, index=X.columns, columns=['select'])
```

#### Cross Validation to decide the best number of Features

```python
from sklearn.model_selection import GridSearchCV, KFold
lir = LinearRegression()
rfe = RFE(lir)

params=[{'n_features_to_select':list(range(1,26))}]
kf = KFold(shuffle=True, n_splits=3, random_state=4)
gs=GridSearchCV(rfe, param_grid=params, scoring='r2', cv=kf)
gs.fit(X, y)
```

```python
gs.best_params_
```

```python
cvr = pd.DataFrame(gs.cv_results_)
cvr.iloc[:,5:]
```

```python
from sklearn.linear_model import LinearRegression
lir = LinearRegression()
rfe = RFE(lir, n_features_to_select=18, verbose=2)
rfe.fit(X, y)
pd.DataFrame(rfe.ranking_, index=X.columns, columns=['select'])
```

### Forward Selection (SFS)

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python
lir = LinearRegression()
rfe=RFE(lir)
sfs1 = sfs(lir, k_features=25, forward=True, scoring='r2', cv=3, verbose=0)
sfs1.fit(X, y)
```

```python
sf = pd.DataFrame(sfs1.subsets_).T
sf
```

```python
sf2=sf[sf['avg_score']==sf['avg_score'].max()]['feature_names']
list(sf2)[0]
```
